볼링에서 점수를 계산하는 프로그램을 짜보시오..

첫 기회때 핀을 점부 쓰러트리면 두번째 기회는 안주고 넘어간다. 점수는 누적 합산됩니다.

핀 쓰러트리는 건 랜덤(0~10)으로.

프레임은 총 10개, 각 프레임은 최대 2번의 투구 기회를 가짐.

스트라이크 (X): 첫 번째 투구에 10핀 전부 쓰러뜨림. → 해당 프레임 점수는 다음 두 투구의 점수를 합산한 값.

스페어 (/): 두 번째 투구까지 합쳐 10핀 전부 쓰러뜨림. → 해당 프레임 점수는 다음 첫 투구 점수를 추가로 합산.

오픈 프레임: 10핀을 모두 쓰러뜨리지 못한 경우 → 단순 합산.

10번째 프레임은 스트라이크/스페어 시 보너스 투구가 주어짐 (최대 3번까지 투구 가능).

핀은 프레임당 최대 10개를 기준으로 계산.

볼링점수 계산 프로그램

볼링은 총 10프레임 = 1게임 으로 진행된다.

프레임 당 최대 2번 던짐 (첫 투구에 스트라이크면 그 프레임은 끝)

스페어 : 두번쨰 투구에서 핀을 다 넘어뜨리면(10점) 다음 투구 1번의 점수 더함

스트라이크 : 처음에 핀을 다 넘어뜨리면 다음 투구 2번의 점수를 더함

마지막프레임에는 예외 규칙 있다..

-> 스트 or 스페어 일때 보너스 투구 (스트 2번 , 스페어 1번) 가능!

! 플레이어가 던져서 나오는 점수는 랜덤으로 처리 할것 !
각 프레임:
첫 투구: 0 ~ 10 중 랜덤
두 번째 투구: 0~(10 - 첫 투구) 중 랜덤

10프레임:
보너스 처리 포함 (스트라이크 or 스페어 시 보너스 투구 2~3번)

```python
import random
import os
import json
```

```python
total_score = 0 # 1게임(10프레임)의 총 점수
Maxscore = 300 # 볼링 최대 점수
MaxbowlingFrame = 10 # 볼링 최대 프레임
strike = 10 # 스트라이크
chance = 2 # 굴리기 횟수 (최대 2회)
totalsave_score_list = [] # 10프레임동안 나온 점수 저장 리스트
startpins = 10
```

```python
def calculate_total_score(frames):
    total = 0  # 누적 점수 저장용 변수

    # 모든 프레임 순회 (최대 10프레임)
    for i in range(len(frames)):
        frame = frames[i]  # 현재 프레임 점수 [10], [4,6] 등

        # 스트라이크인 경우 (한 투구에 10핀)
        if frame == [10]:
            # 다음 두 투구 점수가 있는지 확인
            if i + 1 < len(frames):
                next1 = frames[i + 1]

                # 연속 스트라이크일 경우
                if next1 == [10] and i + 2 < len(frames):
                    next2 = frames[i + 2]
                    bonus = 10 + next2[0]  # 다음 스트라이크 + 다다음 첫 투구
                elif next1 == [10]:
                    bonus = 10  # 다다음 프레임이 없으면 10만 더함
                else:
                    bonus = sum(next1)  # 일반 프레임이면 두 투구 합

                total += 10 + bonus  # 현재 스트라이크 점수 + 보너스
            else:
                total += 10  # 마지막 프레임 등 보너스 불가능한 경우

        # 스페어인 경우 (두 투구 합이 10)
        elif len(frame) == 2 and sum(frame) == 10:
            # 다음 프레임이 있다면 첫 투구 점수를 보너스로 더함
            if i + 1 < len(frames):
                next1 = frames[i + 1]
                bonus = next1[0]
                total += 10 + bonus
            else:
                total += 10  # 마지막 프레임이면 보너스 없이 10점만

        # 일반 프레임인 경우 (합이 10 미만)
        else:
            total += sum(frame)  # 단순히 두 점수 더함

    return total  # 누적 점수 반환
```

```python
Isrunning = True
bowlingFrame = 1 # 시작 볼링 프레임
while(bowlingFrame <= MaxbowlingFrame and Isrunning): # 10프레임까지 계속 가동 10프레임 끝나면 종료
  print("\n현재 프레임은 {} 프레임 입니다.".format(bowlingFrame))

  for a in range(chance):
    isTry = input("공을 굴리겠습니까? 1.네 2.아니오 : ")
    if (isTry == "1"):
      if a == 0:
        first_hit = random.randint(0, 10)
        print("\n첫번째 굴린 결과",first_hit,"이 나왔습니다.")

        # 굴린 결과가 스트라이크면
        if first_hit == strike:
          print("스트라이크!!")
          # 점수 기록
          frame_score = [10]
          totalsave_score_list.append(frame_score)
          # 2번째 기회 안주고 다음 프레임으로 넘기기!
          bowlingFrame += 1  # 다음 프레임으로
          break  # 두 번째 투구 생략하고 for문 탈출

        else:
          remainpins1 = 10 - first_hit
          print("\n현재 남은 핀 : ", remainpins1)

      else: # 두번째 투구
        second_hit = random.randint(0, remainpins1)
        print("\n두 번째 투구 결과 : ",second_hit)
        remainpins2 = remainpins1 - second_hit
        print("현재 남은 핀 : ",remainpins2)

        if a != 0 and first_hit + second_hit == 10:
          print("스페어!")

        total_frame_score = [first_hit, second_hit]
        totalsave_score_list.append(total_frame_score)
        bowlingFrame += 1
        break

    else:
      print("\n굴리기를 취소하셨습니다.")
      startpins = 10
      Isrunning = False
      break

print("볼링 종료!!")
print(totalsave_score_list)
print("총 점수는:", calculate_total_score(totalsave_score_list), "점입니다.")
```
