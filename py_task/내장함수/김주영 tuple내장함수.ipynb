count(x) : 튜플에서 값 x가 몇 번 나오는지 반환

index(x) : 튜플에서 값 x가 처음 나타나는 인덱스를 반환

len() : 요소 갯수 출력

sorted() : 튜플을 정렬한 새 리스트 반환

min() : 튜플에서 가장 작은 값 출력

max() : 튜플에서 가장 큰 값 출력

```python
#count
t_Ex = (1,2,3,3,3,4,5)
count3 = t_Ex.count(3)
count2 = t_Ex.count(2)

print(count3)
print(count2)
```

```python
#index
nums = (3,5,7,9,3,5,7)
print(nums.index(5))
```

```python
#len
nums = (3,5,7,9,3,5,7)
print(len(nums))
```

```python
#sorted
rannums = (3,1,4,2,5,7,6)
print(sorted(rannums))
```

```python
#min
rannums = (3,1,4,2,5,7,6)
print(min(rannums))
```

```python
#max
rannums = (3,1,4,2,5,7,6)
print(max(rannums))
```
