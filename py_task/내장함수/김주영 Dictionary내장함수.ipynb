keys() : 딕셔너리의 **모든 키(key)**를 반환합니다.

values() : 딕셔너리의 **모든 값(value)**을 반환합니다.

items() : 키-값 쌍을 튜플 형태로 반환합니다.

get(key[, default]) : 특정 키의 값을 반환하며, 키가 없을 경우 None 또는 지정한 default를 반환합니다.

update(other) : 다른 딕셔너리의 내용을 업데이트 합니다.

pop(key[, default]) : 특정 키를 제거하고 그 값을 반환합니다. 키가 없으면 default

popitem() : 마지막 (key,value)쌍을 꺼내고 삭제

clear() : 모든 항목을 제거합니다.

copy() : 딕셔너리의 얕은 복사본을 반환합니다.

setdefault(key[, default]) : 키가 존재하면 해당 값을 반환, 없으면 키를 만들고 default 값

fromkeys(seq, value) : 특정 키 특정 값을 가진 딕셔너리 생성

```python
# keys()
m7 = {"Apple": 1,"MSFT": 2,"Google":3,"Amazon":4,"Nvidia":5,"Tesla":6,"Meta":7}
print(m7.keys())
```

```python
#values()
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
print(ABCD.values())
```

```python
#items
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
print(ABCD.items())
```

```python
#get(key[, default]
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
print(ABCD.get("제목"))
print(ABCD.get("저자"))
print(ABCD.get("현재위치"))  # 존재하지 않는 키 (none)
print(ABCD.get("현재위치","집")) # 존재하지 않는 키 (기본값 지정 후)
```

```python
#update
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
ABCD.update({"나이":"26세"}) #새로운 항목 추가
ABCD.update({"제목":"과제수정하기"}) #기존 항목 수정

print(ABCD)
```

```python
#pop
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}

print(ABCD.pop("저자")) # 저자 key 제거후 그 값을 반환
print(ABCD)
```

```python
#popitem
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
testpopitem = ABCD.popitem() #마지막 key value 꺼내기

print(testpopitem)
print(ABCD)
```

```python
#clear
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
ABCD.clear() # 모든요소 삭제

print(ABCD)
```

```python
#copy
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
CopyABCD = ABCD.copy()
CopyABCD["10은"] = 10

print(ABCD)
print(CopyABCD)
```

```python
#setdefault(key[, default])
ABCD = {"제목": "파이썬 과제","저자": "김주영","년도": 2025,"위치": "이태원"}
author = ABCD.setdefault("저자", "이름없음")

print(author) # 출력: 김주영 (기존 키가 존재하면 기존 값 유지)
print(ABCD)  # 딕셔너리는 변경되지 않음
newvalue = ABCD.setdefault("값이없다면", "디폴트로 추가!")
print(newvalue)
print(ABCD)
```

```python
#fromkeys(seq, value)
keys = ["제목","저자","년도","위치"]
New_dict = dict.fromkeys(keys,"과연?")

print(New_dict)
```
