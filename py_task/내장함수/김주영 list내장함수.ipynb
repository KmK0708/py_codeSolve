append(x) : 리스트 끝에 하나의 요소를 추가

extend(iterable) : 리스트에 여러개의 요소들을 추가

insert(i, x) : **원하는 위치(i)**에 요소 추가

remove(x) : 리스트에서 원하는 요소 찾아서 삭제

pop([i]) : 지정한 위치의 요소를 꺼내서 반환. 인덱스를 생략하면 마지막 요소를 꺼냄

index(x) : 리스트에서 값 x의 첫 번째 인덱스 반환

count(x) : 리스트에서 값 x의 개수 반환

sort(reverse=False) : 리스트 오름차순 변경

reverse() : 리스트의 순서를 거꾸로 뒤집음 (정렬 아님)

copy() : 리스트의 얕은 복사본 생성

```python
#append
nums = [1,2,3,4]
nums.append(5)
print(nums)
```

```python
#extend
eats = ["치킨","피자","볶음밥","국수","베이컨"]
eats.extend(["찜닭","닭갈비"])
print(eats)
```

```python
#insert
nums = [1,2,3,4]
nums.insert(3,10)
print(nums)
```

```python
#remove
eats = ["치킨","피자","볶음밥","국수","베이컨"]
eats.remove("치킨")
print(eats)
```

```python
#pop
nums = [1,2,3,4,5,6]
popin = nums.pop()
print(popin)
print(nums)
```

```python
#index

nums = [1,2,3,4,5,6]
print(nums.index(3))
```

```python
#count
face = ["eye","eye","nose","mouth"]
print(face.count("eye"))
```

```python
#sort
rannums = [3,1,5,2,4]
rannums.sort()

print(rannums)

rannums.sort(reverse=True)
print(rannums)
```

```python
#reverse
rannums = [3,1,5,2,4]
rannums.reverse()

print(rannums)
```

```python
#copy
face = ["eye","eye","nose","mouth"]
copy_face = face.copy()

print(face)
print(copy_face)
```
